```python
import numpy as np

np.random.seed(10)
num_samples = 10
dimension = 4
X = np.ceil(10*np.random.random((num_samples,dimension)))
print('X\n',X)
num_classes = 3
W = np.floor(10*np.random.random((dimension,num_classes)))
print('W\n',W)
y = np.ceil((num_classes-1)*np.random.random((num_samples,1)))
y = np.uint8(y)
print('y\n',y)
print()
```

```python
def svm_loss_vectorized(W,X,y,reg):
    s = X.dot(W)
#     print('s\n',s,'\n')
    syi = s[:,y]
#     print('syi\n',syi,'\n')
    syi_again = np.diagonal(syi)
#     print('syi_diagonal_elements\n',syi_again)
#     print(syi_again.shape)
    syi_again = np.reshape(syi_again,(syi_again.shape[1],-1)) # convert row vec to col vec
    sj_syi = s - syi_again
#     print(sj_syi)
    sj_syi_1 = sj_syi + 1
#     print(sj_syi_1)
#     print(list(range(sj_syi_1.shape[0])))
    sj_syi_1[list(range(sj_syi_1.shape[0])),y.T] = 0 # y.T required as y should be a list(so row vector)
#     print(sj_syi_1)
    s2 = np.maximum(0,sj_syi_1)
    print('s2\n',s2,'\n')
    s3 = np.sum(s2,1)  # adding col wise
    s4 = np.sum(s3,0)  # adding row wise
    print(s3,'\n',s4) 
    loss = s4 / num_samples
    print('loss ---> ',loss)
    # add L2 reg. loss
    #     loss += 
# ============================================================================================================================    
    # gradient computation 
    dW = np.zeros_like(W)
    print('Init dW to zero--->\n',dW)
    for i in range(num_samples):
        print('dW\n',dW,'\nX[i,:]\n',X[i,:],'\nX.shape --> ',X.shape,'\n')
        dW += X[i,:][:,np.newaxis]                   #initially add all x's to the gradients
        count = s2[i,s2[i,:]!=0].shape[0]   # go to ith data point and find the non-zero elements, then we can count the shapes
        print('count is: ',count)           
        print('dW after adding all X\'s is-->\n',dW)
        print('y->\n',y)
        dW[list(range(dW.shape[0])),y[i]] -= count * X[i,:]
        print('dW after subtracting the 0 incorrect counts---\n',dW,'\n===========\n')
    dW = dW / num_samples
    print(dW)
    
    
# =============================================================================================================================    
    return loss,dW
#     print(np.sum(s2,1))
    
    
```

```python
svm_loss_vectorized(W,X,y,reg=0)
```
